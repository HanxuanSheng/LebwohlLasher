```python
import numpy as np
import matplotlib.pyplot as plt

#  
x = np.array([50, 200, 500, 1000, 5000, 10000])  # x 
y_python = np.array([0.485704, 1.769425,4.329738,9.079926,45.821759,93.001734])
y_numpy = np.array([ 0.233261,0.896744, 2.242488, 4.477184,23.100689, 47.246823])
x_numba = np.array([50, 200, 500, 1000, 5000, 10000,100000])
y_numba = np.array([1.194981,1.247082,1.326991,1.311458,1.915859,2.257962,16.761750])
y_cython = np.array([0.011905,0.047080,0.115441 ,0.238458, 1.160281,2.296542,24.090066])
y_cython_thread_4 = np.array([0.048010, 0.147860, 0.343214 , 0.643359,3.308643,7.227253, 78.033569])
y_cython_thread_1 = np.array([0.012345 , 0.048640,0.118060 , 0.244946,1.304554,2.528987,24.576735])
y_mpi_4 = np.array([ 0.155765, 0.572392, 1.406733, 2.833825,15.816199, 35.882133])

#  log-log 
plt.figure(figsize=(6, 4))

plt.loglog(x, y_python, label="Python", color='b',marker='*')
plt.loglog(x, y_numpy, label="Numpy", color='g',marker='s')
plt.loglog(x_numba, y_numba, label="Numba+Parallel", color='r',marker='^')
plt.loglog(x_numba, y_cython, label="Cython", color='y',marker='o')
plt.loglog(x_numba, y_cython_thread_4, label="Cython+4 thread", color='k',marker='d')
plt.loglog(x_numba, y_cython_thread_1, label="Cython+1 thread", color='m',marker='p')
plt.loglog(x, y_mpi_4, label="mpi4py 4threads", color='c',marker='.')

plt.xlabel("number of tasks")
plt.ylabel("Time(s)")
plt.legend()
plt.grid(True, which="both", linestyle="--")
plt.title("Log-Log Scaling Plot / size =20*20")
#plt.show()

plt.savefig("size =20")

```
